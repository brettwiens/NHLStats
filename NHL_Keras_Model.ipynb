```python
FinalDataFrame = pd.read_csv("Pre_Model_Dataframe.csv")
FinalDataFrame = FinalDataFrame.drop(['Unnamed: 0'], axis = 1)
# print(FinalDataFrame)
FinalDataFrame['Country'] = np.where(FinalDataFrame['Country'] == "CAN", 1, FinalDataFrame['Country'])
FinalDataFrame['Country'] = np.where(FinalDataFrame['Country'] == "USA", 1, FinalDataFrame['Country'])
FinalDataFrame['Country'] = np.where(FinalDataFrame['Country'] != 1, 0, FinalDataFrame['Country'])

FinalDataFrame['Position'] = np.where(FinalDataFrame['Position'] == "D", 1, FinalDataFrame['Position'])
FinalDataFrame['Position'] = np.where(FinalDataFrame['Position'] != 1, 0, FinalDataFrame['Position'])

print(FinalDataFrame)   
```

```python
     

Targets = FinalDataFrame['Target']
Targets = np.asarray(Targets).astype(np.float32)
FinalDataFrame = FinalDataFrame.drop(['Target','Position','Country'], axis = 1)
FinalDataFrame = np.asarray(FinalDataFrame)
print((FinalDataFrame).dtype)
print(max(Targets))

## Normalization
means = FinalDataFrame.mean()
stdev = FinalDataFrame.std()

normalized_df = (FinalDataFrame - FinalDataFrame.mean())/FinalDataFrame.std()
normalized_df
means.shape

X_train, X_test, Y_train, Y_test = train_test_split(FinalDataFrame, Targets, test_size = 0.2)
```

```python
def coeff_determination(y_true, y_pred):
    SS_res =  K.sum(K.square( y_true-y_pred ))
    SS_tot = K.sum(K.square( y_true - K.mean(y_true) ) )
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )
```

```python
# early_stopping_monitor=EarlyStopping(monitor='val_coeff_determination', patience=5)  #If the model doesn't improve after three epochs, it is probably done.
# mc = ModelCheckpoint('best_model.h5', monitor='val_coeff_determination', mode='max', verbose=1, save_best_only=True)
early_stopping_monitor=EarlyStopping(monitor='val_loss', mode = 'min', patience=5)  #If the model doesn't improve after three epochs, it is probably done.
mc = ModelCheckpoint('best_model.h5', monitor='val_loss', mode='min', verbose=1, save_best_only=True)

seed = 1
#np.random.seed(seed)
#kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
cvscores = []
model = []

#for train, test in kfold.split(train_dataset, target_dataset):
model = Sequential()
model.add(Dense(1024, input_dim=FinalDataFrame.shape[1], activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(512, activation = 'relu'))
model.add(Dense(512, activation = 'tanh'))
model.add(Dense(2056, activation = 'tanh'))
model.add(Dense(2056, activation = 'tanh'))
model.add(Dense(2056, activation = 'relu'))
model.add(Dense(256, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(1,  activation = 'relu'))
model.summary()   

model.compile(loss='mean_absolute_error', optimizer='adam', metrics = ['mean_squared_error', 'mean_absolute_error', coeff_determination])
history = model.fit(X_train, Y_train, validation_split = 0.2, epochs=20000, batch_size=100, verbose=1, callbacks=[mc])

winsound.Beep(440, 100)
winsound.Beep(640, 100)
winsound.Beep(440, 100)
winsound.Beep(880, 100)
winsound.Beep(220, 100)
```
